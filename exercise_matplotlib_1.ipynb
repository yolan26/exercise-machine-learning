Latihan menggunakan library **Matplotlib** untuk visualisasi data. pada contoh dibawah ini menggunakan data penduduk disetiap kelurahan dengan mengkelompokan berdasarkan jumlah laki=laki WNI nya.

menggunakan library **pandas dan numpy** untuk melakukan manipulasi data dari csv untuk mempermudah visualisasi data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/penduduk_gender_head.csv")
table.head()
x_label = table['NAMA KELURAHAN']
plt.bar(x=np.arange(len(x_label)),height=table['LAKI-LAKI WNI'])
plt.show()
```

Selanjutnya agar data lebih mudah dibaca dan dianalisis, kita harus menambahkan label untuk membedakan data disetiap bagian bar nya. dibawah ini kita akan membuat label untuk jumlah penduduk laki-laki di kelurahan Jakarta Pusat.

**label x** adalah seluruh kelurahan di Jakarta Pusat,
**label y** adalah jumlah penduduk laki-laki.

```python
x_label = table['NAMA KELURAHAN']
plt.bar(x=np.arange(len(x_label)),height=table['LAKI-LAKI WNI'])
plt.xticks(np.arange(len(x_label)), table['NAMA KELURAHAN'], rotation=90)
plt.xlabel('Keluarahan di Jakarta pusat')
plt.ylabel('Jumlah Penduduk Laki - Laki')
plt.title('Persebaran Jumlah Penduduk Laki - Laki di Jakarta Pusat')
```

akhirnya didapatkan hasil visualisasi data berbentuk Bar chart untuk mengetahui persebaran jumlah penduduk laki-laki di Jakarta Pusat. terlihat didaerah Tanah tinggi merupakan jumlah penduduk laki-laki terbanyak di Jakarta Pusat, dan Gambir adalah jumlah penduduk laki-laki paling sedikit. selesai
